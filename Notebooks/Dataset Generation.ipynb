```python
import numpy as np
import pandas as pd

def generate_soil_data(n_per_plot=50, n_plots=5):
    np.random.seed(42)
    total_records = n_per_plot * n_plots
    plot_ids = np.repeat([f"Plot {i+1}" for i in range(n_plots)], n_per_plot)

    data = {
        "Plot ID": plot_ids,
        "Soil pH": np.round(np.random.normal(6.5, 0.5, total_records), 2),
        "Organic Matter (%)": np.round(np.random.normal(2.8, 0.6, total_records), 2),
        "Nitrogen (%)": np.round(np.random.normal(0.25, 0.05, total_records), 3),
        "Phosphorus (mg/kg)": np.round(np.random.normal(18, 5, total_records), 1),
        "Potassium (cmol/kg)": np.round(np.random.normal(0.8, 0.2, total_records), 2),
        "Calcium (cmol/kg)": np.round(np.random.normal(10, 3, total_records), 2),
        "Magnesium (cmol/kg)": np.round(np.random.normal(1.5, 0.3, total_records), 2),
        "Sulfur (mg/kg)": np.round(np.random.normal(12, 3, total_records), 1),
        "Zinc (mg/kg)": np.round(np.random.normal(1.2, 0.4, total_records), 2),
        "Iron (mg/kg)": np.round(np.random.normal(100, 20, total_records), 1),
    }
    return pd.DataFrame(data)

soil_df = generate_soil_data(n_per_plot=50, n_plots=5)
sample_df = soil_df.groupby("Plot ID").sample(n=30, random_state=1).reset_index(drop=True)
sample_df.to_csv("Soil Health Data.csv", index=False)
```
